### Random Tensor

Why random tensors?

Random tensors are important because the way neural networks learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data.

'Start with random numbers --> Look at data --> update random numbers --> look at data --> update random numbers'

```python
# Create a random tensor of size (3, 4)
random_tensor = torch.rand(3, 4)
random_tensor
```

```python
random_tensor.ndim
```

### Tensor datatypes

**Note:** Tensor dataypes is one of the 3 big error you'll run into with PyTorch & deep learning
1. Tensors not right datatype
2. Tensor not right shape
3. Tensors not on the right device

Precision in computing - https://en.wikipedia.org/wiki/Precision_(computer_science)#:~:text=In%20computer%20science%2C%20the%20precision,used%20to%20express%20a%20value.

| Type            | Meaning         | Use             |
| --------------- | --------------- | --------------- |
| `torch.float32` | default         | ✅ standard DL   |
| `torch.float16` | lower precision | 🚀 faster (GPU) |
| `torch.float64` | high precision  | 🧪 scientific   |


```python
# Float 32 tensors
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=None, # how detailed your measurement is (e.g. float32 or float16)
                               device = "cpu", # What device is your tensor on
                               requires_grad=False) # whether or not to track gradients/track 'learning' with tensors opertations
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
# .type   used to conver dtype
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
p = float_16_tensor * float_32_tensor
p
```

```python
p.dtype   #Observe:  16bits * 32 bits results 32 bits

## "Promote to higher precision"
# When two tensors have different dtypes: PyTorch automatically converts both to the safer (higher precision) type to avoid information loss
```

```python
int_32_tensor = torch.tensor([3, 6, 9],
                             dtype=torch.long)  #now manupulate this dtype and observe how product happens
int_32_tensor
```

```python
float_32_tensor * int_32_tensor
```

### Two-main Rules to satisfy before matrix-multiplication:
1. The **inner dimensions** must match:
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work
2. **outer dimensions** will be new shape of resulting matrix

### One of the most common errors in deep learning: shape errors

```python
# Shape for matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 8],
                         [9, 10],
                         [11, 12]])

# .matmul() is same as .mm() can use both
torch.matmul(tensor_A, tensor_B)
```

```python
tensor_A.shape, tensor_B.shape
```

To fix tensor shape issues, we can manipulate the shape of one of our tensors using a **transpose**

A **transpose** switches the axes of a given tensor

```python
tensor_B   # 3X2 shape
```

```python
tensor_B.T   # 2X3 shape
```

```python
##Now, the same information is rearranged !!

torch.matmul(tensor_A, tensor_B.T)   # 3X2 * 2X3 = 3X3
```

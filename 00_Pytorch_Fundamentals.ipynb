## Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping --> reshapes an input tensor to a defined shape
* View --> Return a view of an input tensor of certain shape but keep the same memory as the original tensor. But, Works only if memory layout allows.
* Stacking --> combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze --> removes all `1` dimensions from a tensor
* Unsqueeze --> add a `1` dimension to a target tensor
* Permute -->> Return a view of the input with the dimensions permuted (swapped) in a certain way

````python
# create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape, x.dtype, x.ndim
print()
````

````python
# Add an extra dimension
x_reshaped = x.reshape(1,9) # no. of elements must match/valid as 5*2 = 10 and we have 10 total elements
x_reshaped, x_reshaped.shape, x_reshaped.ndim
````

### observe!
Initially, shape of x --> [9] with 1 square bracket  so 1D
but now shape of x_reshaped --> [1,9] with 2 square bracket so 2D

````python
##  More examples of reshape
p = torch.arange(1., 11.)

print(f"Previosu tensor: {p}")
print(f"Previosu shape: {p.shape} ")
print(f"Previosu dimension: {p.ndim}")

# Remove extra dimensions from x_reshaped
p_reshaped = p.reshape(5, 2)  # no. of elements must match/valid as 5*2 = 10 and we have 10 total elements
print(f"\nNew tensor: {p_reshaped}")
print(f"New shape: {p_reshaped.shape}")
print(f"New dimesnsion: {p_reshaped.ndim}")
````

 Observe the shape change and Brackets here !

> intitally, [10] with 1 bracket so 1D but now [5, 2] with 2 brackets so 2D

## Pytorch tensors and Numy

Numpy is a popular scientific Python numerical library

And because of this, PyTorch has functionality to interact to it.

* Data in Numpy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PTorch tensor -> Numpy -> `torch.Tensor.numpy()`

````python
# Numpy array to Tensor

import torch
import numpy as np

array = np.arange(1.0, 8.0)   # np.arange !!
tensor = torch.from_numpy(array) # warning: when converting from numpy -> pytorch, pytorch reflects numpy's default datatype of float64 unless specified otherwise !!
array, array.dtype, tensor
````

## Reporducibility (trying to take random out of random)

In short how a NN learns:

`start with random numbers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> again...`

To reduce the randomness in NN and PyTorch comes the concept of a **random seed**

Essentially what the random seed does is "flavour" the randomness.

````python
torch.rand(3, 3)
````

````python
import torch

#create two random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
````

#### What is a “seed”?

* Seed = starting point for randomness

Think:

* Random numbers are not truly random
* They follow a hidden pattern
* Seed decides where to start in that pattern

````python
# Let's make some random but reproducible tensors
import torch

#set the random seed
RANDOM_SEED = 42    # Means: Start randomness from point 42

torch.manual_seed(RANDOM_SEED)        ##   !!
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)        ##   !!
random_tensor_D = torch.rand(3, 4)

# Same seed -> same starting point
# Same operations -> same numbers


print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
````

#### What is **Reproducibility** ?

`Getting the same results again and again`

#### Why it matters in ML:
* Debugging
* Experiments
* Comparing models


#### Real-world analogy
Think:

* Random = shuffled cards
* Seed = fixed shuffle pattern

Same seed -> same shuffle every time

####  Golden rules
1. torch.rand() → random every time
2. torch.manual_seed(x) → fixes randomness
3. Same seed + same code → same output

#### One-line summary

Seed controls randomness → ensures reproducibility

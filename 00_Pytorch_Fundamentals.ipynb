### Random Tensor

Why random tensors?

Random tensors are important because the way neural networks learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data.

'Start with random numbers --> Look at data --> update random numbers --> look at data --> update random numbers'

| `torch.float32` | default         | ✅ standard DL   |

### One of the most common errors in deep learning: shape errors

````python
# Shape for matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 8],
                         [9, 10],
                         [11, 12]])

# .matmul() is same as .mm() can use both
torch.matmul(tensor_A, tensor_B)
````

````python
tensor_A.shape, tensor_B.shape
````

To fix tensor shape issues, we can manipulate the shape of one of our tensors using a **transpose**

A **transpose** switches the axes of a given tensor

````python
tensor_B   # 3X2 shape
````

````python
tensor_B.T   # 2X3 shape
````

````python
##Now, the same information is rearranged !!

torch.matmul(tensor_A, tensor_B.T)   # 3X2 * 2X3 = 3X3
````

## Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping --> reshapes an input tensor to a defined shape
* View --> Return a view of an input tensor of certain shape but keep the same memory as the original tensor. But, Works only if memory layout allows.
* Stacking --> combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze --> removes all `1` dimensions from a tensor
* Unsqueeze --> add a `1` dimension to a target tensor
* Permute -->> Return a view of the input with the dimensions permuted (swapped) in a certain way

````python
# create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape, x.dtype, x.ndim
print()
````

````python
# Add an extra dimension
x_reshaped = x.reshape(1,9) # no. of elements must match/valid as 5*2 = 10 and we have 10 total elements
x_reshaped, x_reshaped.shape, x_reshaped.ndim
````

### observe!
Initially, shape of x --> [9] with 1 square bracket  so 1D
but now shape of x_reshaped --> [1,9] with 2 square bracket so 2D

````python
##  More examples of reshape
p = torch.arange(1., 11.)

print(f"Previosu tensor: {p}")
print(f"Previosu shape: {p.shape} ")
print(f"Previosu dimension: {p.ndim}")

# Remove extra dimensions from x_reshaped
p_reshaped = p.reshape(5, 2)  # no. of elements must match/valid as 5*2 = 10 and we have 10 total elements
print(f"\nNew tensor: {p_reshaped}")
print(f"New shape: {p_reshaped.shape}")
print(f"New dimesnsion: {p_reshaped.ndim}")
````

 Observe the shape change and Brackets here !

> intitally, [10] with 1 bracket so 1D but now [5, 2] with 2 brackets so 2D

````python
# Change the view
z = x.view(1,9)
z, z.shape
````

````python
# Changing z changes x ( because a view of tensor shares the same memory as the original input)
z[:,0] = 5
z, x
````

````python
# Stack tensors om top pf each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
````

````python
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
````

````python
# vstack-> dim=0 and hstack-> dim=1
````

````python
# Squeeze
# torch.squeeze() = remove all single dimensions from a target tensor
print(f"Previosu tensor: {x_reshaped}")
print(f"Previosu shape: {x_reshaped.shape} ")
print(f"Previosu dimension: {x_reshaped.ndim}")

# Remove extra dimensions from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
print(f"New dimesnsion: {x_squeezed.ndim}")


## Obseerve: Dimension Change
````

````python
# Unsqueeze

## torch.unsqueeze() = adds a single dimension to a target tensor at a specific dim
print(f"Previous target: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")
print(f"Previous Dimension: {x_squeezed.ndim}")

# Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
print(f"New Dimension: {x_unsqueezed.ndim}")
````

````python
# Permute
## torch.permute(input, dims) = reaaranges the dimesnions of a target tensor in a specified order. returns view.

#* permute mostly used in Image stuffs
x_original = torch.rand(size= (224, 224, 3))  # H, W, #channels

#Permute the original tensor to reaarange the axis (or dim) order
x_permuted = x_original.permute (2, 0, 1) # shits axix 0->1, 1->2, 2->0

print(f"Previous Shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}") # [color channels, H, W]
````
